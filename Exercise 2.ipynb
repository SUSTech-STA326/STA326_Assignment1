```python
# Imports
import requests  # send request
from bs4 import BeautifulSoup  # parse web pages
import pandas as pd  # csv
from time import sleep  # wait
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
```

## 1.Find all web-page websites of the 100 recent paper

```python
url = 'https://arxiv.org/list/cs/pastweek?skip=0&show=100' # Directly use the web-page showing the first 100
page = requests.get(url)
# page = requests.get(url)
soup = BeautifulSoup(page.text)
```

```python
soup
```

```python
## Center on content
soup2 = soup.find_all('div', id = "content")
soup2[0]
```

```python
## Center on list-identifier
soup3 = soup2[0].find_all('span', class_ = "list-identifier")
soup3
```

```python
## take out the websites
locations = []
for s in soup3:
    locations.append(s.find('a')['href'])
print(locations)
```

## 2.Take out detailed information

### Go though all 100 papers

```python
titles = []
subjects = []
authors = []
abstracts = []
for location in locations:
    url = 'https://arxiv.org' + location # Directly use the web-page showing the first 100
    page = requests.get(url)
    # page = requests.get(url)
    soup = BeautifulSoup(page.text)
    informations = soup.find_all('div', id = 'abs')[0]
    # title
    title = informations.find('h1', class_ = "title mathjax").text
    titles.append(title)
    print(title)
    # subject
    subject = informations.find('td', class_ = "tablecell subjects").text
    subjects.append(subject)
    # author
    author = informations.find('div', class_ = "authors").text
    authors.append(author)
    # abstract
    abstract = informations.find('blockquote', class_ = "abstract mathjax").text
    abstracts.append(abstract)
titles
```

### data cleaning

```python
title_clean = []
for i in range(len(titles)):
    title_clean.append(titles[i][6:])
```

```python
title_clean
```

```python
subject_clean = []
for i in range(len(subjects)):
    subject_clean.append(subjects[i][1:])
```

```python
subject_clean
```

```python
author_clean = []
for i in range(len(authors)):
    author_clean.append(authors[i][8:])
author_clean
```

```python
abstract_clean = []
for i in range(len(abstracts)):
    abstract_clean.append(abstracts[i][10:])
abstract_clean
```

## 3.save as a csv file

```python
csv_name =  "Exercise2.csv"
# csv_dir = "./output"
# # YOUR CODE HERE
data_list = pd.DataFrame({'Title':title_clean,'Subject':subject_clean
                          ,'Author':author_clean,'Abstract':abstract_clean})
data_list.to_csv(csv_name,index = False)
data_list.head()
# raise NotImplementedError()
```
